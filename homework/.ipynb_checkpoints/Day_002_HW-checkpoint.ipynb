透過索引值做排序，axis 可以指定第幾欄，ascending 用於設定升冪或降密  

```python
app_train.sort_index(axis=0,ascending=False)
```

```python
app_train.sort_values(by ="AMT_INCOME_TOTAL")
```

判斷哪些店名是遺失值

```python
print(app_train[app_train.loc[:,"OWN_CAR_AGE"].isnull()])
```

判斷哪些市場規模不是遺失值

```python
print(app_train[app_train.loc[:,"OWN_CAR_AGE"].notnull()])
```

有遺失值的觀測值都刪除  

```python
drop_value = app_train.dropna()
print(drop_value)
```

有遺失值的觀測值填補 0  

```python
filled_value = app_train.fillna(0)
print(filled_value)
```
